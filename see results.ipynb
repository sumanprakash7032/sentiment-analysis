```python
import pandas as pd
import re
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('final.csv')
len(df.sentiment)
```

```python
#plot different count plot for different mobiles 
for i in range(len(df.sentiment)):
    q=df.sentiment[i]
    p=re.sub("[^a-zA-Z,]", " ", q)
    p=p.split(',')
    sns.countplot(p)
    plt.title(df.title[i])
    plt.show()
```
